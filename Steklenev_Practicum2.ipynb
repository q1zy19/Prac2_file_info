Постройте гистограмму из 20 наиболее употребительных слов в приложенном файле, а таже сделайте сводку по файлу включая количество слов, строк и символов (с пробелами и без них).

```python
with open('Уэллс_-_Война_миров.txt', encoding='windows-1251') as file:
    f=file.read()
    lines = f.splitlines()
text = ' '.join(f.replace('\n', ' ').replace('\xa0', ' ').lower().split())
```

```python
import re

text = re.sub(r'[^\w\s]', '', text);text
```

```python
from collections import Counter
```

```python
Counter(text)
```

```python
Counter(text.replace(" ",""))
```

```python
len(lines)
```

```python
words = text.split()
word_counts = Counter(words); word_counts
```

```python
top_words = dict(word_counts.most_common(20));top_words
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(x=[i for i in top_words.keys()], y=[j for j in top_words.values()])
```
